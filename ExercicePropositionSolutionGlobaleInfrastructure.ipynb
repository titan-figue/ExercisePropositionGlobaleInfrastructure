## 2. Mise en place d'un environnement technique adapté

Dans ce premier audit on a pu faire une recommandation préliminaire. La plupart des outils recommandées seront décrit à continuation :

### 2.1. IDE Integrated Developpement Environnement

Un IDE un environnement de programmation complet qui se présente sous la forme d'une application.

Il permet de s'assurer que les développeurs ont les bons outils pour être plus efficaces. 

Il est composé généralement d'un éditeur de code, d'un compilateur, d'un débogueur et d'un générateur d'interface graphique. Ils existent plusieurs IDE dans le cadre de l'intégration continue

On propose de travailler avec Visual Studio code parce que il s'adapte à la plus part des langages de programmation.

### 2.2. Integration et Livraison Continue CI/CD

L'integration et la livraison continue des logiciels c'est un processus complexe qui requiere plus que simplement appuyer sur un bouton ou deux. 

Pour assurer que les versions d'application soient suivies et bien testés avant d'être mises en production, C'est recommandable d'utiliser des outils que permettront l'orchestration de l'intégration et de la livraison continue (CI/CD) appropiées. 

Aujourd'hui, existent plusieurs platformes qui permettent de travailler avec, Atlassian Jira, Atlassian Bamboo, GitLab, AzureDevOps, Jenkins X Pipelines, circleCI, Tuleap, etc.

A continuation vous verrez celles qu'on as regardé dans une première révision.

Atlassian Jira, c'est un excellent approche agile tres agreable pour planifier et developper son produit. #### 2.2.2. Atlassian Bamboo

Atlassian Bamboo est un serveur de CI/CD qui peut fournir:

* Automatise la gestion des versions des applications logicielles. avec un marquage des versions.
* La construction et les test automatisés du code source.
* Des mise à jour sur les builds qui ont réussi et échoués.
* Des outils de reporting pour l'analyse statistique.
* Crée des pipelines de livraison continue pour partager des artifacts plans buildés dans un environnement de déploiement. 
* Les déclencheurs sont créés en fonction des modifications détectées dans le référentiel. Pousse les notifications de Bitbucket, un calendrier défini, l'achèvement d'une autre build ou toute combinaison de ceux-ci.




Vous pouvez prendre un instantané de vos artefacts depuis une compilation réussie dans une version.

Ensuite, vous pouvez déployer le projet, annuler et/ou promouver les versions dans les environnements de development, de staging ou de production.

Suivez quoi, quand et qui a déployé des versions dans vos environnements.

##### 4.2.3 Dockerfile

Dockerfile sert pour construire des images, le Dockerfile on l'écris par example sur vim (linux) en suivant quelques commandes de ceux qui se trouvent à continuation:

•  ADD copie les fichiers d'une source sur l'hôte dans le propre système de fichiers du conteneur à la destination définie.

•  CMD peut être utilisé pour exécuter une commande spécifique dans le conteneur.

•  ENTRYPOINT définit une application par défaut à utiliser chaque fois qu'un conteneur est créé avec l'image.

•  ENV définit les variables d'environnement.

•  EXPOSE associe un port spécifique pour permettre la mise en réseau entre le conteneur et le monde extérieur.

•  FROM définit l'image de base utilisée pour démarrer le processus de construction.

•  MAINTAINER définit un nom complet et une adresse e-mail du créateur d'image.

•  RUN est la directive d'exécution centrale pour Dockerfiles.

•  USER définit l'UID (ou nom d'utilisateur) qui doit exécuter le conteneur.

•  VOLUME est utilisé pour permettre l'accès du conteneur à un répertoire sur la machine hôte.

•  WORKDIR définit le chemin où la commande, définie avec CMD, doit être exécutée.

•  LABEL vous permet d'ajouter une étiquette à votre image de menu fixe.

##### 4.2.6 Docker Swarm Mode

Le mode swarm indique à Docker que vous exécuterez de nombreux moteurs Docker et que vous souhaitez coordonner les opérations sur tous. Le mode Swarm combine la capacité non seulement de définir l'architecture de l'application, comme Docker Compose, mais aussi de définir et de maintenir des niveaux de haute disponibilité, la mise à l'échelle, l'équilibrage de charge, etc. Avec toutes ces fonctionnalités, le mode Swarm est utilisé plus souvent dans les environnements de production.

### 5. Atlassian bamboo-server

Pour creer une session du bamboo-server, c'est bien de creer un conteneur dans lequel on va ouvrir un Docker image. 

1. Créer un repertoire host.
2. Dans le terminal ubuntu utiliser un volume Docker:

$> docker volume create --name bambooVolume

$> docker run -v bambooVolume:/var/atlassian/application-data/bamboo --name="bamboo" --init -d -p 54663:54663 -p 8085:8085 atlassian/bamboo-server

3. Demarrer Atlassian Bamboo:

$> docker run -v /data/bamboo:/var/atlassian/application-data/bamboo --name="bamboo-server" --host=bamboo-server --init -d -p 54663:54663 -p 8085:8085 atlassian/bamboo-server

Et bamboo est disponible dans: http://localhost:8085*.

4. En bamboo créer un nouveau projet par example "DevOps data-retriever", dans ce nouveau projet vous creez un plan (pipeline), click en [Create plan]

5. Ensuite, vous configurez ce plan et clone le repository a partir de votre repository host, dans notre example on a utlisé GIT puis on as cloné avec HTTPS depuis GitHub.
    Click en [Test connection] pour creer la connection entre bamboo et GitHub.
    Click en [Configure plan].

6. Apres avoir crée le plan, on configure le "Job" et on cree les "task".
    Selectionner par example Malven. Maven c’est un “exécutable” c’est un programme externe que Bamboo utilise dans la phase de building. Généralement cet exécutable compile le code source et génère des fichiers compilés exécutables.
    * Click en [Add task].
    * Add new exécutable.
    * Puis click en [save].



    
    
 7. Des que le projet est configuré on peut commencer a tourner les builders.

### 6. Droit informatique

Tout contrat emis doit passer par les main d'un experte juriste. 

Cependant, Il faut être au courant de ce que le contrat doit definir par example: 

   - Le droit d'acteur.
   - Vie privée (RGPD).
   - Liberté d'expression.
   - Droit à l'image.
   - La licence de logiciel (droit d'usage => claire et precis)
   - Le contrat de maintenance (Est-ce qu'il va contenir: la maintenance corrective, la maintenance évolutive, la maintenance réglamentaire et la maintenance préventive).
   - La licence de logiciel libre (libre circulation de code souce).
   - Le contrat de développement de logiciel spécifique.
   - Les obligations des parties.
   - La clause de propriété intellectuelle.
   
Le type de contrat peut être: 
   - Le contrat d'integration de logiciels.
   - Le contrat d'outsourcing.
   - Le contrat ASP "Application Service Provider"
   - Le contrat SAAS "Software As A Service"
   - Le contrat d'hébergement.
   - Le contrat de conseil en informatique (AMOA)
   - Le contrat de référencement
   - Les contrats d'entiercement ou de séquestre
